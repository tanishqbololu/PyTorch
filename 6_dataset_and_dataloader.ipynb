```python
from sklearn.datasets import make_classification
import torch
```

```python
# Systhetic classification dataset using sklearn
X, y = make_classification(
    n_samples=10,  # Total number of data points
    n_features=2,  # Number of input columns per sample
    n_informative=2,  # How many features actually contain useful information for prediction
    n_redundant=0,  # duplicate/correlated features
    n_classes=2,  # Number of output categories
    random_state=42 #Fixes randomness
)
```

```python
# Convert the data to PyTorch Tensors
X = torch.tensor(X, dtype=torch.float32)  #X = torch.from_numpy(X).float()
y = torch.tensor(y, dtype=torch.float32)
```

```python
X,y
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CustomDataset(Dataset):

  def __init__(self, features, labels):
    self.features = features
    self.labels = labels

  def __len__(self):
    return self.features.shape[0]  # no. of rows in features

  def __getitem__(self, index):
    return self.features[index], self.labels[index]
```

```python
dataset = CustomDataset(X, y)
```

```python
len(dataset)
```

```python
dataset[0]
```

```python
dataset[2]
```

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=True, num_workers=2)
```

```python
import time
for batch_features, batch_labels in dataloader:
  print(batch_features)
  print(batch_labels)
  print("-"*50)
```
