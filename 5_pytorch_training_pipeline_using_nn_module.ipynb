# Libraries and Dataset

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=["id","Unnamed: 32"], inplace=True)
df.head()
```

# Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:,0], test_size=0.2)
```

# Scaling

```python
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
X_train
```

# Label Encoding

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

# NumPy arrays to PyTorch tensors

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

# Defining the model

*   We will make a simple NN with one hidden neuron.
*   But as our X has 30 features means we will have 30 weights from the input layer to hidden layer



```python
X_train.dtype, y_train.dtype
```

The model weigts are in float32, so we must either convert training data to floar32 with .float() or double the model with .double() to make it float64.
But doubling the model will make it occupy more space in memory unnecessarily.

```python
X_train_tensor = X_train_tensor.float()
y_train_tensor = y_train_tensor.float()
```

```python
import torch.nn as nn
```

```python
class MySimpleNN(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out

  def loss_function(self, y_pred, y):  # y = true labels -> (train/val/test)
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7    # 1 * 10^-7 --> 0.0000001
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)

    loss = -(y * torch.log(y_pred) + (1-y) * torch.log(1-y_pred)).mean()
    return loss
```

# Important Parameters

```python
learning_rate = 0.1
epochs = 25
```

```python
# define loss function
loss_function = nn.BCELoss()  # Binary Cross Entropy Loss
```

# Training Pipeline

```python
# create model
model = MySimpleNN(X_train_tensor.shape[1])

# define optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# Define Loop
for epoch in range(epochs):

  # Forward Pass
  y_pred = model(X_train_tensor)
  #print(y_pred)

  # Zero Gradients --> Before running new epoch make gradients=0
  # model.linear.weight.grad.zero_()
  # model.linear.bias.grad.zero_()
  optimizer.zero_grad()  #automatically makes all grads zero

  # Loss Calculation
  #loss = model.loss_function(y_pred, y_train_tensor)
  loss = loss_function(y_pred, y_train_tensor.view(-1,1))  # This is loss fucntion made with nn.BCELoss()
  #print(loss)

  # Backward Pass
  loss.backward()

  # Parameters Update -> update weights & bias:
  # with torch.no_grad():
  #   model.linear.weight -= learning_rate * model.linear.weight.grad   # w_new = w_old - n.dL/dw_old
  #   model.linear.bias -= learning_rate * model.linear.bias.grad       # b_new = b_old - n.dL/db_old

  optimizer.step()  # -> this will automatically update the weights and bias

  # Print loss in each epoch
  print(f"Epoch: {epoch}, Loss: {loss.item()}") # .item() converts a scalar tensor into a Python float and removes it from the computation graph, mainly for logging or monitoring.”
```

# Evaluation

```python
X_test_tensor = X_test_tensor.float()
y_test_tensor = y_test_tensor.float()
```

```python
with torch.no_grad():
  y_pred = model(X_test_tensor)
# But to calculate accuracy we must make them into 0s and 1s and compare with y_test.
  y_pred = (y_pred > 0.5).float()
# accuracy
accuracy = (y_pred == y_test_tensor).float().mean()
print(f"Accuracy: {accuracy.item()}")
```
